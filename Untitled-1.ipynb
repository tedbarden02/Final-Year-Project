```python
import pandas as pd
import numpy as np
from sdv.single_table import GaussianCopulaSynthesizer
from sdv.metadata import SingleTableMetadata
import matplotlib.pyplot as plt

# Step 1: Load the Dataset
try:
    data = pd.read_csv('imputed.csv')
except FileNotFoundError:
    print("Error: 'imputed.csv' not found. Please ensure the file is in the correct directory.")
    exit(1)

# Step 2: Check for Missing Values
print("Missing values in original data:")
print(data.isnull().sum())
if data.isnull().any().any():
    print("Warning: Missing values detected. Filling with 0 for numerical columns.")
    data = data.fillna(0)

# Step 3: Set Up Metadata
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data)
# Explicitly set column types
metadata.update_column('Date/Time', sdtype='datetime')
metadata.update_column('Date/time end', sdtype='datetime')
metadata.update_column('Size fraction', sdtype='categorical')
numerical_cols = [
    'Mass v [µg/m**3]', 'Na+ [µg/m**3]', '[NH4]+ [µg/m**3]', 'K+ [µg/m**3]', 
    'Mg2+ [µg/m**3]', 'Ca2+ [µg/m**3]', 'Cl- [µg/m**3]', '[NO3]- [µg/m**3]', 
    '[SO4]2- [µg/m**3]', '[C2O4]2- [µg/m**3]', 'Br- [µg/m**3]', 'C org [µg/m**3]', 
    'EC [µg/m**3]', 'TC [µg/m**3]', 'Ca [ng/m**3]', 'Ti [ng/m**3]', 'V [ng/m**3]', 
    'Cr [ng/m**3]', 'Mn [ng/m**3]', 'Fe [ng/m**3]', 'Ni [ng/m**3]', 'Cu [ng/m**3]', 
    'Zn [ng/m**3]', 'Pb [ng/m**3]', 'As [ng/m**3]', 'Se [ng/m**3]', 'Sr [ng/m**3]', 
    'Rb [ng/m**3]', 'Ba [ng/m**3]', 'La [ng/m**3]', 'Ce [ng/m**3]'
]
for col in numerical_cols:
    if col in data.columns:
        metadata.update_column(column_name=col, sdtype='numerical')
    else:
        print(f"Warning: Column {col} not found in dataset.")

# Step 4: Initialize and Fit the Gaussian Copula Model
model = GaussianCopulaSynthesizer(metadata=metadata)
model.fit(data)

# Step 5: Generate Synthetic Data
synthetic_data = model.sample(num_rows=1000)


# Step 7: Save Synthetic Data
synthetic_data.to_csv('synthetic_data_gaussian_copula.csv', index=False)

# Step 8: Validate Synthetic Data
# Statistical comparison
valid_cols = [col for col in numerical_cols if col in data.columns and col in synthetic_data.columns]
print("\nOriginal Data Statistics:")
print(data[valid_cols].describe())
print("\nSynthetic Data Statistics:")


# Visualize distributions for key columns
key_cols = ['Na+ [µg/m**3]', 'Mass v [µg/m**3]', 'Fe [ng/m**3]']
for col in key_cols:
    if col in data.columns and col in synthetic_data.columns:
        plt.figure(figsize=(10, 5))
        plt.hist(data[col], bins=30, alpha=0.5, label='Original', color='#4e79a7')
        plt.hist(synthetic_data[col], bins=30, alpha=0.5, label='Synthetic', color='#f28e2b')
        plt.title(f'Distribution Comparison for {col}')
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()

print("Synthetic data generation complete. Saved to 'synthetic_data_gaussian_copula.csv'.")
```

```python
import pandas as pd
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata

# Load the original dataset
data = pd.read_csv('imputed.csv')

# Convert date columns to datetime
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
data['Date/time end'] = pd.to_datetime(data['Date/time end'])

# Compute duration in seconds
data['Duration'] = (data['Date/time end'] - data['Date/Time']).dt.total_seconds()

# Drop the end date column
data = data.drop(columns=['Date/time end'])

# Create and update metadata
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data)
metadata.update_column(column_name='Date/Time', sdtype='datetime')
metadata.update_column(column_name='Duration', sdtype='numerical')

# Initialize and train the CTGAN model
model = CTGANSynthesizer(metadata=metadata)
model.fit(data)

# Generate synthetic data
synthetic_data_ctgan = model.sample(num_rows=1000)

# Ensure duration is non-negative
synthetic_data_ctgan['Duration'] = synthetic_data_ctgan['Duration'].clip(lower=0)

# Compute "Date/time end" in synthetic data
synthetic_data_ctgan['Date/time end'] = synthetic_data_ctgan['Date/Time'] + pd.to_timedelta(synthetic_data_ctgan['Duration'], unit='s')

# Save the corrected synthetic data
synthetic_data_ctgan.to_csv('corrected_synthetic_data.csv', index=False)
print("Corrected synthetic data saved to 'corrected_synthetic_data.csv'.")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
import pymc as pm
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# Load the dataset
data = pd.read_csv('data.csv')

# Convert Date/Time to numerical (days since first date)
data['Date/Time'] = pd.to_datetime(data['Date/Time'], errors='coerce')
data['Date/Time_num'] = (data['Date/Time'] - data['Date/Time'].min()).dt.total_seconds() / (60 * 60 * 24)
data['Date/Time_num'] = data['Date/Time_num'].fillna(0)  # Handle NaNs

# Create dummy variables for Size fraction
size_dummies = pd.get_dummies(data['Size fraction'], prefix='size', dummy_na=False).astype(float)

# Construct and scale predictor matrix X
X_df = pd.concat([size_dummies, data[['Date/Time_num']]], axis=1).astype(float)
scaler = StandardScaler()
X = scaler.fit_transform(X_df)
X = np.nan_to_num(X, nan=0.0)  # Replace any remaining NaNs

# Define censored columns (numerical columns starting from the 5th)
censored_cols = [
    'Na+ [µg/m**3]', '[NH4]+ [µg/m**3]', 'K+ [µg/m**3]', 'Mg2+ [µg/m**3]', 'Ca2+ [µg/m**3]',
    'Cl- [µg/m**3]', '[NO3]- [µg/m**3]', '[SO4]2- [µg/m**3]', '[C2O4]2- [µg/m**3]', 'Br- [µg/m**3]',
    'C org [µg/m**3]', 'EC [µg/m**3]', 'TC [µg/m**3]', 'Ca [ng/m**3]', 'Ti [ng/m**3]',
    'V [ng/m**3]', 'Cr [ng/m**3]', 'Mn [ng/m**3]', 'Fe [ng/m**3]', 'Ni [ng/m**3]',
    'Cu [ng/m**3]', 'Zn [ng/m**3]', 'Pb [ng/m**3]', 'As [ng/m**3]', 'Se [ng/m**3]',
    'Sr [ng/m**3]', 'Rb [ng/m**3]', 'Ba [ng/m**3]', 'La [ng/m**3]', 'Ce [ng/m**3]'
]

# Filter valid columns
censored_cols = [col for col in censored_cols if col in data.columns]

# Function to extract detection limit
def get_detection_limit(series):
    censored = series[series.astype(str).str.startswith('<')]
    if len(censored) > 0:
        d_str = censored.iloc[0].lstrip('<').strip()
        try:
            return float(d_str)
        except ValueError:
            return None
    return None

# Simple imputation fallback
def simple_impute(series, d_limit):
    series = series.copy()
    censored = series.astype(str).str.startswith('<')
    series[censored] = d_limit / 2 if d_limit else 0
    return pd.to_numeric(series, errors='coerce').fillna(0)

# Impute censored values using Bayesian method with fallback
imputed_data = data.copy()
d_limits = {}
for col in censored_cols:
    d = get_detection_limit(data[col])
    d_limits[col] = d
    y = data[col].copy()
    censored = y.astype(str).str.startswith('<')
    
    if d is not None and censored.any():
        y[censored] = d  # Replace censored values with detection limit
        y = pd.to_numeric(y, errors='coerce')
        if y.isna().any():
            print(f"Warning: Non-numeric values in {col}. Filling NaNs with {d}.")
            y = y.fillna(d)
        
        # Ensure non-censored values are not below detection limit
        y[~censored] = y[~censored].clip(lower=d)
        
        try:
            with pm.Model() as model:
                # Initialize beta with linear regression estimates
                lr = LinearRegression().fit(X, y)
                beta_init = lr.coef_
                beta = pm.Normal('beta', mu=beta_init, sigma=1, shape=X.shape[1])
                sigma = pm.HalfNormal('sigma', sigma=np.std(y) if np.std(y) > 0 else 1)
                mu = pm.math.dot(X, beta)
                observed = pm.Censored('observed', pm.Normal.dist(mu=mu, sigma=sigma), lower=d, observed=y.values)
                trace = pm.sample(1000, tune=1000, return_inferencedata=False, progressbar=True, initvals={'beta': beta_init})
            
            imputed_values = np.mean(trace['observed'], axis=0)
            imputed_data[col] = np.maximum(imputed_values, 0)
        except Exception as e:
            print(f"Bayesian imputation failed for {col}: {e}. Using simple imputation.")
            imputed_data[col] = simple_impute(data[col], d)
    else:
        print(f"No censored values in {col}. Using original data.")
        y = pd.to_numeric(y, errors='coerce')
        if y.isna().any():
            print(f"Warning: Non-numeric values in {col}. Filling NaNs with 0.")
            y = y.fillna(0)
        imputed_data[col] = y

# Prepare data for CTGAN
non_censored_cols = ['No', 'Date/Time', 'Size fraction', 'Altitude [m]', 'Mass v [µg/m**3]']
non_censored_cols = [col for col in non_censored_cols if col in data.columns]
ctgan_data = imputed_data[non_censored_cols + censored_cols]

# Create metadata for CTGAN
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(ctgan_data)
metadata.update_column(column_name='Date/Time', sdtype='datetime')
metadata.update_column(column_name='Size fraction', sdtype='categorical')
for col in censored_cols:
    metadata.update_column(column_name=col, sdtype='numerical')

# Train CTGAN
model = CTGANSynthesizer(metadata=metadata, epochs=300, verbose=True)
model.fit(ctgan_data)

# Generate 1000 synthetic samples
synthetic_data = model.sample(num_rows=1000)

# Reapply censoring to synthetic data
for col in censored_cols:
    d = d_limits.get(col)
    if d is not None:
        synthetic_data[col] = synthetic_data[col].apply(lambda x: f'<{d}' if x < d else x)

# Save synthetic data
synthetic_data.to_csv('synthetic_data_bayesian.csv', index=False)
print("Synthetic data with Bayesian imputation saved to 'synthetic_data_bayesian.csv'.")

# Validate statistical properties
def convert_censored(series, d_limit):
    series = series.copy()
    masked = series.astype(str).str.startswith('<')
    series[masked] = d_limit if d_limit is not None else 0
    return pd.to_numeric(series, errors='coerce').fillna(0)

original_stats = pd.DataFrame()
synthetic_stats = pd.DataFrame()
for col in censored_cols:
    d = d_limits.get(col)
    original_stats[col] = convert_censored(data[col], d).describe()
    synthetic_stats[col] = convert_censored(synthetic_data[col], d).describe()

print("\nOriginal data statistics:")
print(original_stats)
print("\nSynthetic data statistics:")
print(synthetic_stats)
```

```python
#Removing whitespaces and "<"
def calculate_percentage(col):
    col_str = col.astype(str).str.strip()  
    count = col_str.str.startswith("<").sum()  
    total = col.notna().sum()  
    return (count / total) * 100 if total > 0 else 0  # Calculate percentage, handle division by zero


cols = data.iloc[:, 7:]

# Apply the function to each column
percentages = cols.apply(calculate_percentage)

# Create a DataFrame for presentation
result = percentages.reset_index()
result.columns = ['Column', 'Percentage']

# Print the results with two decimal places
for idx, row in result.iterrows():
    print(f"{row['Column']}: {row['Percentage']:.2f}%")
```
